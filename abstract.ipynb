```python
parking_lots = 20

class Vehicle:
    def __init__(self):
        pass

    def park(self, parking_lots):
        if parking_lots <= 0:
            print("No parking space available")
        else:
            parking_lots -= 1
            print("Vehicle parked")

    def unpark(self, parking_lots):
        parking_lots += 1
        print("Vehicle unparked")     
        return parking_lots
    
class Car(Vehicle):
    def __init__(self):
        super().__init__()
        
class Bike(Vehicle):
    def __init__(self):
        super().__init__()   

class Truck(Vehicle):
    def __init__(self):
        super().__init__()     

def get_vehicle(vehicle_type):
    if vehicle_type == "car":
        return Car()
    elif vehicle_type == "bike":
        return Bike()
    elif vehicle_type == "truck":
        return Truck()
    else:
        print("Unknown vehicle type")

while True:
    print("\nParking Lot Management System")
    print("Welcome to the parking lot,")
    print("Please select your purpose:")
    print("1. Park")
    print("2. Unpark")
    print("3. Status")
    print("4. Exit")  
    option = input("Enter option (1-4): ")
    if option == '1':
        vehicle_type = input("Enter vehicle type (car/bike/truck): ").lower()
        selected_vehicle = get_vehicle(vehicle_type)
        selected_vehicle.park(parking_lots)
    elif option == '2':
        vehicle_type = input("Enter vehicle type (car/bike/truck): ").lower()
        selected_vehicle = get_vehicle(vehicle_type)
        selected_vehicle.unpark(parking_lots)
    elif option == '3':
        print(f"Available parking lots: {parking_lots}")
    elif option == '4':
        print("Exiting the system.")                                                              
        break
    else:
        print("Invalid option. Please try again.")

print("Thank you for using the Parking Lot Management System.")
```

```python
parking_lots = 20
list_history = []
amount_history = 0

class Vehicle:
    def __init__(self):
        pass

    def park(self, parking_lots):
        
        if parking_lots <= 0:
            print("No parking space available")
        else:
            parking_lots -= 1
            print("Vehicle parked. Remaining lots:", parking_lots)
            list_history.append("Parked a vehicle \n")
            return parking_lots

    def unpark(self, parking_lots):
        parking_lots += 1
        print("Vehicle unparked. Remaining lots:", parking_lots)
        list_history.append("Unparked a vehicle /n")
        return parking_lots


class Car(Vehicle):
    def __init__(self):
        super().__init__()

class Bike(Vehicle):
    def __init__(self):
        super().__init__()

class Truck(Vehicle):
    def __init__(self):
        super().__init__()


def get_vehicle(vehicle_type):
    if vehicle_type == "car":
        return Car()
    elif vehicle_type == "bike":
        return Bike()
    elif vehicle_type == "truck":
        return Truck()
    else:
        print("Unknown vehicle type")
        return None


while True:
    print("\nParking Lot Management System")
    print("1. Park")
    print("2. Unpark")
    print("3. Status")
    print("4. History")
    print("5. Exit")  

    option = input("Enter option (1-4): ")

    if option == '1':
        vehicle_type = input("Enter vehicle type (car/bike/truck): ").lower()
        selected_vehicle = get_vehicle(vehicle_type)
        if selected_vehicle:
           parking_lots = selected_vehicle.park(parking_lots)
           amount_history += 10

    elif option == '2':
        vehicle_type = input("Enter vehicle type (car/bike/truck): ").lower()
        selected_vehicle = get_vehicle(vehicle_type)
        if selected_vehicle:
           parking_lots = selected_vehicle.unpark(parking_lots)

    elif option == '3':
        print(f"Available parking lots: {parking_lots}")

    elif option == '4':
        print("History of parking activities:", list_history)
        print("Total amount collected:", amount_history)

    elif option == '5':
        print("Exiting the system.")
        break

    else:
        print("Invalid option. Please try again.")

print("Thank you for using the Parking Lot Management System.")
```
